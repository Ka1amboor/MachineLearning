2. Предложите способ заменить отсутствующие значения из пункта 1 и реализуйте его. Способ нужно аргументировать(сказать почему это достаточно хорошее приближение)

````python
import pandas as pd


df = pd.read_csv('russian-housing-market.csv')

valid_ratio = df[df['life_sq'].notna()]['life_sq'] / df[df['life_sq'].notna()]['full_sq']
mean_ratio = valid_ratio.mean()

df['life_sq'] = df['life_sq'].fillna(df['full_sq'] * mean_ratio)

df['life_sq'] = df['life_sq'].fillna(df['life_sq'].median())

#корреляция между life_sq и full_sq, жилая площадь обычно составляет определённую долю от общей площади, использование среднего отношения позволяет сохранить эту пропорцию
#устойчивость к выбросам, среднее отношение менее чувтсвительно к выбросам чем абсолютные значения площади, тк оно нормализует данные
````

1. Как менялась средняя цена квартиры в Нагатинском Затоне(Nagatinski Zaton) с течением времени

````python
import pandas as pd
import matplotlib.pyplot as plt


try:
    df = pd.read_csv('russian-housing-market.csv', sep=',', parse_dates=['timestamp'])
except FileNotFoundError:
    print("Ошибка: Файл не найден.")
    exit()


nagatinski_zaton = df[df['sub_area'].str.contains('Nagatinskij Zaton', case=False, na=False)]

print(f"Найдено записей: {len(nagatinski_zaton)}")
if len(nagatinski_zaton) == 0:
    print("Данные отсутствуют.")
    exit()


monthly_prices = nagatinski_zaton.resample('MS', on='timestamp')['price_doc'].mean().reset_index()


monthly_prices['price_doc'] = monthly_prices['price_doc'].interpolate()


plt.figure(figsize=(12, 6))
plt.plot(monthly_prices['timestamp'], monthly_prices['price_doc'], marker='o', linestyle='-', color='blue')
plt.title('Средняя цена в Нагатинском Затоне')
plt.xlabel('Дата')
plt.ylabel('Цена (руб.)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

5. Визуализируйте зависимость средней стоимости квартир в зависимости от этажа

````python
import pandas as pd
import matplotlib.pyplot as plt


try:
    df = pd.read_csv('russian-housing-market.csv', sep=',')
except FileNotFoundError:
    print("Ошибка: Файл не найден.")
    exit()


required_columns = ['floor', 'price_doc']
if not set(required_columns).issubset(df.columns):
    print("Ошибка: Отсутствуют необходимые столбцы.")
    exit()


df_clean = df[
    (df['floor'] > 0) &
    (df['price_doc'] > 0) &
    (df['floor'] <= df['max_floor'])
].dropna(subset=['floor', 'price_doc'])


floor_price = df_clean.groupby('floor')['price_doc'].agg(['mean', 'count']).reset_index()
floor_price.columns = ['Этаж', 'Средняя_цена', 'Количество_квартир']


floor_price = floor_price[floor_price['Количество_квартир'] >= 10]


plt.figure(figsize=(12, 6))
plt.plot(
    floor_price['Этаж'],
    floor_price['Средняя_цена'],
    marker='o',
    linestyle='-',
    color='blue',
    label='Средняя цена'
)


plt.title('Зависимость стоимости квартир от этажа')
plt.xlabel('Этаж')
plt.ylabel('Средняя цена (руб.)')
plt.grid(True, alpha=0.3)
plt.xticks(range(1, int(floor_price['Этаж'].max()) + 1))


for i, row in floor_price.iterrows():
    plt.text(
        row['Этаж'],
        row['Средняя_цена'],
        f"{row['Количество_квартир']}",
        ha='center',
        va='bottom'
    )

plt.legend()
plt.tight_layout()
plt.show()
````

7. Постройте график изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года*

\* Задание на доп. балл

````python
import pandas as pd
import matplotlib.pyplot as plt


try:
    df = pd.read_csv('russian-housing-market.csv', sep=',', parse_dates=['timestamp'])
except FileNotFoundError:
    print("Ошибка: Файл не найден.")
    exit()

required_columns = ['timestamp', 'price_doc', 'full_sq']
if not set(required_columns).issubset(df.columns):
    print("Ошибка: Отсутствуют столбцы 'timestamp', 'price_doc' или 'full_sq'.")
    exit()


df_filtered = df[
    (df['full_sq'] < 50) &
    (df['price_doc'] > 0) &
    (df['timestamp'].between('2011-01-01', '2014-12-31'))
].dropna(subset=['price_doc', 'full_sq', 'timestamp'])


if df_filtered.empty:
    print("Нет данных для построения графика.")
    exit()


monthly_median = df_filtered.resample('ME', on='timestamp')['price_doc'].median()

plt.figure(figsize=(12, 6))
plt.plot(
    monthly_median.index,
    monthly_median.values,
    marker='o',
    linestyle='-',
    color='blue',
    label='Медианная цена'
)
plt.title('Изменение медианной цены квартир <50 кв.м (2011–2014)')
plt.xlabel('Дата')
plt.ylabel('Цена (руб.)')
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
````
